```python
import requests
from langchain.document_loaders import TextLoader
```

```python
url = "https://raw.githubusercontent.com/hwchase17/chat-your-data/refs/heads/master/state_of_the_union.txt"
```

```python
res=requests.get(url)
```

```python
print(res.text)
```

```python
with open("state_of_the_union.txt","w") as f:
  f.write(res.text)
```

```python
loder=TextLoader('/content/state_of_the_union.txt')
```

```python
document=loder.load()
```

```python
print(document[0].page_content)
```

```python
from langchain.text_splitter import CharacterTextSplitter
```

```python
text_splitter=CharacterTextSplitter(chunk_size=500,chunk_overlap=50)
```

```python
chunks=text_splitter.split_documents(document)
```

```python
print(chunks[3].page_content)
```

```python
from langchain.embeddings import OpenAIEmbeddings
```

```python
from weaviate import Client
from langchain.vectorstores import Weaviate
```

```python
from weaviate.embedded import EmbeddedOptions
import os
client = weaviate.WeaviateClient(
    embedded_options=EmbeddedOptions()
)
```

```python
from google.colab import userdata
OPENAI_API_KEY=userdata.get('OPENAI_API_KEY')
```

```python
!pip install langchain_weaviate
```

```python
from langchain_weaviate.vectorstores import WeaviateVectorStore
```

```python
client.connect()
```

```python
vectorstore = WeaviateVectorStore.from_documents(
    client=client,
    documents=chunks,
    embedding=OpenAIEmbeddings(openai_api_key=OPENAI_API_KEY),
    by_text=False
)
```

```python
retriever=vectorstore.as_retriever()
```

```python
from langchain.prompts import ChatPromptTemplate
```

```python
template="""You are an assistant for question-answering tasks.
Use the following pieces of retrieved context to answer the question.
If you don't know the answer, just say that you don't know.
Use ten sentences maximum and keep the answer concise.
Question: {question}
Context: {context}
Answer:
"""
```

```python
prompt=ChatPromptTemplate.from_template(template)
```

```python
print(prompt)
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser
```

```python
llm_model=ChatOpenAI(openai_api_key=OPENAI_API_KEY,model_name="gpt-3.5-turbo")
```

```python
query="Why Ukraine wants to join NATO?"
```

```python
output_parser = StrOutputParser()
```

```python
rag_chain = (
    {"context": retriever,  "question": RunnablePassthrough()}
    | prompt
    | llm_model
    | output_parser
)
```

```python
rag_chain.invoke(query)
```
